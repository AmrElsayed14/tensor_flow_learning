### Getting information from tensors 
when dealing with tensors you probably want to be aware of the following attribuites:
* shape
* Rank 
* Axis or dimension
* Size

```python
# Create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements aling 0 axis:", rank_4_tensor.shape[0])
print("Elements along last axis",rank_4_tensor.shape[-1])
print("Total number of elements in the tensor",tf.size(rank_4_tensor).numpy())
```

**Matrix Multiplication**

In machine learning matrix multiplication is one of the common tensor operations.

```python
# Matrix multiplication in tensorflow
tf.matmul(tensor,tensor)
```

```python
# Matrix multiplication with Python operator "@"
tensor @ tensor
```

```python
# Create a tensor (3,2) tensor
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
# Create a tensor (3,2) tensor
Y = tf.constant([[7,8],
                 [9,10],
                 [11,12]])
X,Y
```

```python
#X @ tf.reshape(Y,shape=(2,3))
tf.matmul(X,tf.reshape(Y,shape=(2,3)))
```

```python
#Can do the same with transpose
X,tf.transpose(X), tf.reshape(X,shape=(2,3))
```

```python
# try matrix multiplication with transpose rather than reshape 
tf.matmul(tf.transpose(X),Y)
```

**The dot product**
Matrix multiplication is also referred to as the dor product.
You can perform matrix multiplication using:
* `tf.matmul()`
* `tf.tensordot()`

```python
# perform the dot product on X and Y (requres X or Y to be transposed)
tf.tensordot(tf.transpose(X),Y,axes=1)
```

```python
# Perfomr matrix multiplication between X and Y (trasnposed) 
tf.matmul(X,tf.transpose(Y))
```

```python
# Perfomr matrix multiplication between X and Y (reshaped) 
tf.matmul(X,tf.reshape(Y,shape=(2,3)))
```

```python
# check the values of Y, and reshape Y and transposed Y
print("Normal Y;")
print(Y,"\n")

print("Y reshaped to (2,3);")
print(tf.reshape(Y,(2,3)),"\n")

print("Y transposed: ")
print(tf.transpose(Y),"\n")
```

Generally, when you perform matrix multiplication on 2 tensors and one of the axis doesn't line up, you will transpose rather than reshape one of the tensors to get satisfy  the matrix multiplication rules.

### Aggregating tensors
 Aggregating tensors = condensing them from multiple values down to a smaller amount of values.


```python
# Get the absolute values 
D = tf.constant([-7,-10])
tf.abs(D)
```

lets go through the following forms of aggregation:
* Get the min* Get the max.
* Getthe mean of tensor
* Get the sum of tensor

```python
# Creating a random tesnorwith values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0,100,size=50))
tf.size(E),E.shape, E.ndim
```

```python
# Find the minimium and max
tf.reduce_min(E)
tf.reduce_max(E)
# Find the mean
tf.reduce_mean(E)
tf.reduce_sum(E)
```

```python
# Find the variance of our tensor 
import tensorflow_probability as tfp
tfp.stats.variance(E)
tf.math.reduce_variance(tf.cast(E,tf.float32)) # both ways are right
```

```python
# standard deviation
tf.math.reduce_std(tf.cast(E,tf.float32))
```

### Find the positional maximium and minimium

```python
# Create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the maximum position
max = tf.argmax(F)
max, F[max]
```

```python
# Find the minimum position
min = tf.argmin(F)
min, F[min]
```

```python
assert tf.reduce_min(F) == F[min] # using assert to throw eror if they are not equal otherwise no output
```
